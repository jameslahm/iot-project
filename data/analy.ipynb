```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

## Message

### Distance

```python
df = pd.read_csv('message-distance.csv')
df=df.rename(columns={'Unnamed: 0':'distance'})
df['avg'] =(df['1']+df['2']+df['3'])/3
df['pkg'] = (((df['1']==100)*1  + (df['2']==100)*1 + (df['3']==100)*1)/3)*100
```

```python
df['1'] = 100-df['1']
df['2'] = 100-df['2']
df['3'] = 100-df['3']
df['avg'] = 100 - df['avg']
df['pkg'] = 100 - df['pkg']
```

```python
df
```

```python
fig = plt.figure(figsize=(12, 6))
ax = plt.gca()
x = np.arange(len(df))
rects=plt.bar(x, 100-df['avg'], width=0.5)
plt.ylabel("Bits Error Ratio Avg")
plt.ylim(0,80)
plt.xticks(x,df['distance'])
plt.xlabel("Distance(cm)")
for rect in rects:
    height = rect.get_height()
    ax.annotate('{:.2f}%'.format(height),
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')
plt.savefig('message_distance_bit.png')
```

```python
fig = plt.figure(figsize=(12, 6))
ax = plt.gca()
x = np.arange(len(df))
rects=plt.bar(x, 100-df['pkg'], width=0.5)
plt.ylabel("Packet Error Ratio")
plt.ylim(0,120)
plt.xticks(x,df['distance'])
plt.xlabel("Distance(cm)")
for rect in rects:
    height = rect.get_height()
    ax.annotate('{:.2f}%'.format(height),
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')
plt.savefig('message_distance_pkg')
```

## Location

### Distance

```python
df = pd.read_csv('location-distance.csv')
df=df.rename(columns={'Unnamed: 0':'distance'})
df['avg'] =(df['1']+df['2']+df['3'])/3 * 331 / 2 / 1000 * 100
df['std'] = (df.iloc[:,1:4] * 331 / 2 / 1000 * 100).std(axis=1)
```

```python
fig = plt.figure(figsize=(12, 6))
ax = plt.gca()
x = np.arange(len(df))
rects=plt.bar(x, df['avg'], width=0.5)
plt.ylabel("Location Distance Avg(cm)")
plt.ylim(0,200)
plt.xticks(x,df['distance'])
plt.xlabel("Distance(cm)")
for rect in rects:
    height = rect.get_height()
    ax.annotate('{:.2f}'.format(height),
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')
plt.savefig('location_distance_avg.png')
```

```python
fig = plt.figure(figsize=(12, 6))
ax = plt.gca()
x = np.arange(len(df))
rects=plt.bar(x, df['std'], width=0.5)
plt.ylabel("Location Distance Std")
plt.ylim(0,20)
plt.xticks(x,df['distance'])
plt.xlabel("Distance(cm)")
for rect in rects:
    height = rect.get_height()
    ax.annotate('{:.2f}'.format(height),
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')
plt.savefig('location_distance_std.png')
```
